```python
DUBLIN_BIKES=pd.read_csv("dublin_bikes.csv")
#DUBLIN_BIKES.head()
#DUBLIN_BIKES.tail()
#DUBLIN_BIKES.columns

WORKING_DAY=pd.read_csv("No_Holiday_bikes.csv")

DUBLIN_BIKES_2=pd.read_csv("dublin.csv")
```

This API provides real time information on the Dublin bikes.This times series data has a total of 111 rows and 14 columns and it shows the real time dublin bike usage over a period of time.Nevertheless,I noticeed the table has some irrelavant column names in the DataFrame such as "Unnamed" and "Name(which contains same variables as address)".Therefore, this data would undergo cleaning so as to work with the relavant DataFrame for the analysis.

```python
DUBLIN_BIKES_REALTIME=pd.DataFrame(DUBLIN_BIKES,columns=["NUMBER","CONTRACT NAME", "ADDRESS","POSITION",
                    "BANKING","BONUS",'BIKE STANDS',
                   'AVAILABLE BIKE STANDS','STATUS','LAST UPDATE',"AVAILABLE BIKES",
                   "DATES"])
#DUBLIN_BIKES_REALTIME
#DUBLIN_BIKES_REALTIME.head()

DUBLIN_BIKE_STATIC=DUBLIN_BIKES_2.rename(columns= {"Number":"NUMBER", "Address":"ADDRESS","Latitude":"LATITUDE","Longitude":"LONGITUDE"})

DUBLINBIKE_STATIC=pd.DataFrame(DUBLIN_BIKE_STATIC,columns=["NUMBER","ADDRESS","LATITUDE","LONGITUDE"])
#DUBLINBIKE_STATIC
```

Dublin Bikes are in two kinds which is the real time and static data.Static data provides stable information
like station position,longitude and latitude etc.Fotunately, Static data can be downloaded manually in CSV file format or 
accessedthrough the API. Dynamic data which is our real time data are refreshed every minute and can be accessed only through 
the API(Dublinbikes DCC - Dublinbikes 2021/04/01-2021/07/01 - data.gov.ie, 2022). 
Futhermore,After cleaning, we can use the dataset to get answers for the questions stated in the hypothesis. 
H0: The availability of bikes in dublin does not depend on location. H1: The availability of bikes in dublin depends on location.
To get access to the locations using latitude and longitude we therefore augment the realtime data with a static data so as to
get additional columns for the latitude and longitude which was scrapped from the web. 

The next step, having the static data we augment with the Dublin bike data.This can be merged with a "outer join" 

```python
BIKE_DATA=pd.merge(DUBLIN_BIKES_REALTIME,DUBLIN_BIKE_STATIC, how="outer")
BIKE_DATA
```

Therefore, with the updated bike data, we now have a combination of both the static and dynamic(real time bike) data.Therefore,
there is an increase in the size of the dataset which we would be working with in the course of this project.

```python
BANK_HOLIDAY=pd.DataFrame(BIKE_DATA,columns=["NUMBER","CONTRACT NAME", "ADDRESS","POSITION",
                    "BANKING","BONUS",'BIKE STANDS',
                   'AVAILABLE BIKE STANDS','STATUS','LAST UPDATE',"AVAILABLE BIKES",
                   "DATES","LATITUDE","LONGITUDE"])
BANK_HOLIDAY

#BANK_HOLIDAY.info()


```

```python
WORKING_DAYY=pd.DataFrame(WORKING_DAY,columns=["NUMBER","CONTRACT NAME", "ADDRESS","POSITION",
                    "BANKING","BONUS",'BIKE STANDS',
                   'AVAILABLE BIKE STANDS','STATUS','LAST UPDATE',"AVAILABLE BIKES",
                   "DATES"])
WORKING_DAYY
```
